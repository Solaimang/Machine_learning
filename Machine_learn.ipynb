```python
income_order = ['low','medium','high']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enco = OrdinalEncoder(categories=[income_order])
```

```python
enco.fit(df[['income']])
```

```python
final = pd.DataFrame(enco.transform(df[['income']]))
```

```python
final
```

```python
df
```

```python
df['income']=final
```
